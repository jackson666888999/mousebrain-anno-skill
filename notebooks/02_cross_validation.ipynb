# Tutorial 2: Multi-Tool Cross-Validation

Compare annotations from multiple methods and compute consensus.

```python
import scanpy as sc
from mbanno import ConsensusAnnotator
from pathlib import Path

adata = sc.read_h5ad("data/query.h5ad")
print(f"Cells: {adata.n_obs:,}")
```

```python
annotator = ConsensusAnnotator(
    reference="allen-consensus-wmb",
    taxonomy_level="subclass"
)

# Run multiple methods for cross-validation
results = annotator.annotate(
    adata,
    methods=["marker"],  # add "mapmycells", "scanvi", "celltypist", "singler"
    min_confidence=0.5,
)
```

```python
# Compare agreement
annotations = results["annotations"]
label_cols = [c for c in annotations.columns if c.endswith("_label")]

for col in label_cols:
    n_types = annotations[col].nunique()
    print(f"{col}: {n_types} unique types")
```

```python
# Confidence breakdown
conf = results["confidence"]
print("Confidence levels:")
print(conf["confidence_level"].value_counts())
```

```python
annotator.save_results(results, Path("results/tutorial_02"))
annotator.generate_report(results, Path("results/tutorial_02"))
```
