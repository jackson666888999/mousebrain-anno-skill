# Tutorial 3: Spatial Data Annotation

Annotate MERFISH, Slide-seq, or STARmap spatial data with region constraints.

```python
import scanpy as sc
from mbanno import ConsensusAnnotator
from pathlib import Path

# Load spatial data
adata = sc.read_h5ad("data/merfish_sample.h5ad")
print(f"Cells: {adata.n_obs:,}, Genes: {adata.n_vars:,}")
```

```python
annotator = ConsensusAnnotator(
    reference="allen-consensus-wmb",
    taxonomy_level="subclass"
)

# Region constraint: cortex
results = annotator.annotate(
    adata,
    methods=["marker"],
    region_constraint="cortex",
    min_confidence=0.5,
)
```

```python
annotations = results["annotations"]
print(f"Top cell types in cortex:")
print(annotations["consensus_label"].value_counts().head(10))
```

```python
annotator.save_results(results, Path("results/tutorial_03"))
```
